#### 1. How likely is it that you roll doubles when rolling two dice?
Probability of doubles = 0.1682



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(52)
```

```python
## Set the number of trials

n_trials = 10_000
n_rows = 10_000

## Roll two dice

n_dice = 2
n_cols = 2

# Use np's random choice
# Select from [1,2,3,4,5,6]

rolls = np.random.choice([1,2,3,4,5,6], n_trials * n_dice).reshape(n_rows, n_cols)


rolls
```

```python
# programmatically recognize doubles:
doubles = []

for row in rolls:
    if row [0] == row [1]:
        doubles.append(row)
        
doubles
```

#### 5. Compare Heights
- Men have an average height of 178 cm and standard deviation of 8cm.
- Women have a mean of 170, sd = 6cm.
- Since you have means and standard deviations, you can use np.random.normal to generate observations.
- If a man and woman are chosen at random, what is the likelihood the woman is taller than the man?
Modeled probability that random woman is taller than random man 0.2058


```python
# number of sample pairs we want to generate
n5_rows = 10_000  

# one random man's height based on given mu and sigma
# one random woman's height based on given mu and sigma
m = pd.DataFrame(np.random.normal(loc=178, scale=8, size =(n5_rows, 1)))
w = pd.DataFrame(np.random.normal(loc=170, scale=6, size =(n5_rows, 1)))

# compare the two values across the two DataFrames:
bools = w > m
bools
prob_woman_taller = bools.mean()
prob_woman_taller
```

#### 7. There's a 70% chance on any given day that there will be at least one food truck at Travis Park. However, you haven't seen a food truck there in 3 days. How unlikely is this?
#### How likely is it that a food truck will show up sometime this week?



```python
# What are the outcomes?
#0 - no truck
#1 - truck is present

trucks = np.random.choice([0, 1], size = (10000, 3), p = [0.3, 0.7])
#probability that no truck shows up
(trucks.sum(axis = 1) == 0).mean()
```

```python
# truck shows up at least once.
(trucks.sum(axis = 1) >= 1).mean()
```

```python
# Food truck shows up at least once in a week
weekly_trucks = np.random.choice([0, 1], size = (10000, 7), p = [0.3, 0.7])
(weekly_trucks.sum(axis=1) > 0).mean()
```

#### 8. If 23 people are in the same room, what are the odds that two of them share a birthday? What if it's 20 people? 40?
23 people = 0.36639
20 people = 0.32414
40 people = 0.26043
These numbers are for exactly 2 people sharing a birthday, which explains why the probability goes down for 40 people.



```python
num_people_in_room = 40   # columns
num_possible_birthdays = 366  # range of outcomes
num_trials = 10**5 # rows

birthday_sets_a = np.random.choice(
    range(1, 367), size = (num_trials, num_people_in_room))
birthday_sets = pd.DataFrame(birthday_sets_a)
birthday_sets
two_share_bday = (birthday_sets.nunique(axis=1) == (num_people_in_room - 1))
two_share_bday
p_two_share_bday = two_share_bday.mean()
p_two_share_bday
```
